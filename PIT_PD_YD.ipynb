```python
# Creating function for lag
```

```python
def lag1(df,col):
    df[col+"_lag1"] = df[col].shift(1)
    
    df[col+"_lag1"+"_LQ1"]= df[col+"_lag1"].shift(1)
    df[col+"_lag1"+"_LQ2"]= df[col+"_lag1"+"_LQ1"].shift(1)
    df[col+"_lag1"+"_LQ3"]= df[col+"_lag1"+"_LQ2"].shift(1)
    df[col+"_lag1"+"_LQ4"]= df[col+"_lag1"+"_LQ3"].shift(1)
    return df
```

```python
def lag2(df,col):
    df[col+"_lag2"] = df[col].shift(2)
    
    df[col+"_lag2"+"_LQ1"]= df[col+"_lag2"].shift(1)
    df[col+"_lag2"+"_LQ2"]= df[col+"_lag2"+"_LQ1"].shift(1)
    df[col+"_lag2"+"_LQ3"]= df[col+"_lag2"+"_LQ2"].shift(1)
    df[col+"_lag2"+"_LQ4"]= df[col+"_lag2"+"_LQ3"].shift(1)
    return df
```

```python
def lag4(df,col):
    df[col+"_lag4"] = df[col].shift(4)
    
    df[col+"_lag4"+"_LQ1"]= df[col+"_lag4"].shift(1)
    df[col+"_lag4"+"_LQ2"]= df[col+"_lag4"+"_LQ1"].shift(1)
    df[col+"_lag4"+"_LQ3"]= df[col+"_lag4"+"_LQ2"].shift(1)
    df[col+"_lag4"+"_LQ4"]= df[col+"_lag4"+"_LQ3"].shift(1)
    return df
```

```python
# Creating function for log
```

```python
def log(df,col):
    #df[col] =df[col].astype(float)
    df[col+"_log"] = np.log10(df[col])
    
    df[col+"_log"+"_LQ1"]= df[col+"_log"].shift(1)
    df[col+"_log"+"_LQ2"]= df[col+"_log"+"_LQ1"].shift(1)
    df[col+"_log"+"_LQ3"]= df[col+"_log"+"_LQ2"].shift(1)
    df[col+"_log"+"_LQ4"]= df[col+"_log"+"_LQ3"].shift(1)
    
    return df
```

```python
def log_diff1(df,col):
    
    #Temp
    df[col+"_log"+"_temp"] = np.log10(df[col])
    
    df[col+"_log_lag1"] =  df[col+"_log"+"_temp"].shift(-1)
    df[col+"_log_diff1"] = df[col+"_log_lag1"] - df[col+"_log"+"_temp"]
    df[col+"_log_diff1"] = df[col+"_log_diff1"].shift(1)
    df =df.drop([col+"_log_lag1"],axis=1)
    df =df.drop([col+"_log"+"_temp"],axis=1)
    
    
    df[col+"_log_diff1"+"_LQ1"]= df[col+"_log_diff1"].shift(1)
    df[col+"_log_diff1"+"_LQ2"]= df[col+"_log_diff1"+"_LQ1"].shift(1)
    df[col+"_log_diff1"+"_LQ3"]= df[col+"_log_diff1"+"_LQ2"].shift(1)
    df[col+"_log_diff1"+"_LQ4"]= df[col+"_log_diff1"+"_LQ3"].shift(1)
    
    
    return df
```

```python
def log_diff2(df,col):
    
    #Temp
    df[col+"_log"+"_temp"] = np.log10(df[col])
    
    df[col+"_log_lag2"] =  df[col+"_log"+"_temp"].shift(-2)
    df[col+"_log_diff2"] = df[col+"_log_lag2"] - df[col+"_log"+"_temp"] 
    df[col+"_log_diff2"] = df[col+"_log_diff2"].shift(2)
    df =df.drop([col+"_log_lag2"],axis=1)
    df =df.drop([col+"_log"+"_temp"],axis=1)
    
    df[col+"_log_diff2"+"_LQ1"]= df[col+"_log_diff2"].shift(1)
    df[col+"_log_diff2"+"_LQ2"]= df[col+"_log_diff2"+"_LQ1"].shift(1)
    df[col+"_log_diff2"+"_LQ3"]= df[col+"_log_diff2"+"_LQ2"].shift(1)
    df[col+"_log_diff2"+"_LQ4"]= df[col+"_log_diff2"+"_LQ3"].shift(1)
    
    return df
```

```python
def log_diff4(df,col):
    
    #Temp
    df[col+"_log"+"_temp"] = np.log10(df[col])  
    
    df[col+"_log_lag4"] =  df[col+"_log"+"_temp"].shift(-4)
    df[col+"_log_diff4"] = df[col+"_log_lag4"] -  df[col+"_log"+"_temp"]
    df[col+"_log_diff4"] = df[col+"_log_diff4"].shift(4)
    df =df.drop([col+"_log_lag4"],axis=1)
    df =df.drop([col+"_log"+"_temp"],axis=1)
    
    df[col+"_log_diff4"+"_LQ1"]= df[col+"_log_diff4"].shift(1)
    df[col+"_log_diff4"+"_LQ2"]= df[col+"_log_diff4"+"_LQ1"].shift(1)
    df[col+"_log_diff4"+"_LQ3"]= df[col+"_log_diff4"+"_LQ2"].shift(1)
    df[col+"_log_diff4"+"_LQ4"]= df[col+"_log_diff4"+"_LQ3"].shift(1)
    
    return df
```

```python
def growth1(df,col):
    
    # Temp 
    df[col+"_lag1"+"_temp"] = df[col].shift(1)
    
    df[col+"_growth1"] =(df[col]- df[col+"_lag1"+"_temp"])/df[col+"_lag1"+"_temp"]
    
    df =df.drop([col+"_lag1"+"_temp"],axis=1)
    
    df[col+"_growth1"+"_LQ1"]= df[col+"_growth1"].shift(1)
    df[col+"_growth1"+"_LQ2"]= df[col+"_growth1"+"_LQ1"].shift(1)
    df[col+"_growth1"+"_LQ3"]= df[col+"_growth1"+"_LQ2"].shift(1)
    df[col+"_growth1"+"_LQ4"]= df[col+"_growth1"+"_LQ3"].shift(1)
    
    
    return df
```

```python
# Load Statsmodels 
import statsmodels.api as sm

# ADF Test
from statsmodels.tsa.stattools import adfuller

# KPSS Test
from statsmodels.tsa.stattools import kpss
```

```python
def stationary_test(df,col):
    dftest = adfuller(df[col],maxlag=1,regression='n',autolag=None)
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    dfoutput=dfoutput.to_frame(name = col)
    dfoutput1 = dfoutput.transpose()
    dfoutput1.rename(columns = {"Test Statistic": "ADF_Test_Statistic",
                             "p-value": "ADF_p-value",
                             "#Lags Used": "ADF_#Lags_Used",
                             "Number of Observations Used": "ADF_Number_of_Observations_Used",
                             "Critical Value (1%)": "ADF_Critical_Value_(1%)",
                             "Critical Value (5%)": "ADF_Critical_Value_(5%)",
                             "Critical Value (10%)": "ADF_Critical_Value_(10%)"
                              }, inplace = True)
    kpsstest = kpss(df[col],nlags=2, regression='c')
    kpss_output = pd.Series(kpsstest[0:3], index=['Test Statistic','p-value','#Lags Used'])
    for key,value in kpsstest[3].items():
        kpss_output['Critical Value (%s)'%key] = value
    kpss_output=kpss_output.to_frame(name = col) 
    kpss_output1 = kpss_output.transpose()
    kpss_output1.rename(columns = {"Test Statistic": "KPSS_Test_Statistic",
                               "p-value": "KPSS_p-value",
                               "#Lags Used": "KPSS_#Lags_Used",
                               "Critical Value (10%)": "KPSS_Critical_Value_(10%)",
                               "Critical Value (5%)": "KPSS_Critical_Value_(5%)",
                               "Critical Value (2.5%)": "KPSS_Critical_Value_(2.5%)",
                               "Critical Value (1%)": "KPSS_Critical_Value_(1%)"
                              }, inplace = True)
    ADF_KPSS = dfoutput1.join(kpss_output1)

    return ADF_KPSS
```

```python
# Importing libraries for multivariate test 

from sklearn import metrics
from scipy import stats
import statsmodels.formula.api as smf
from statsmodels.compat import lzip
import statsmodels.stats.api as sms
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.api import add_constant
import numpy as np
import scipy.stats as st
```

```python
def multivariate_test_2(train_df,test_df,train_y,test_y,var1,var2):
    X_train = train_df[[var1,var2]]
    X_test = test_df[[var1,var2]]
    X2_train = sm.add_constant(X_train)
    X2_test = sm.add_constant(X_test)
    est = sm.OLS(train_y, X2_train)
    est2 = est.fit()
    model_results = pd.DataFrame()
    model_results["Variable1"] =[var1]
    model_results["Variable2"] =[var2]
    model_results["VIF"] = [1/(1-est2.rsquared)]
    #r = np.corrcoef(X.iloc[:, 0], X.iloc[:, 1])
    #model_results["correlation"] = r[0,1]
    r = X_train.corr()
    model_results["correlation"] = r.iloc[0,1]
    model_results["Model_Intercept"] = [est2.params[0]]
    model_results["Model_coeff_Variable1"] =[est2.params[1]]
    model_results["Model_coeff_Variable2"] = [est2.params[2]]
    
    confidenceLevel = 0.99           #99% CI given
    degrees_freedom = len(X_train)-1 
    sampleMean_V1 = np.mean(X_train.iloc[:,0]) #sample mean
    sampleStandardError_V1 = st.sem(X_train.iloc[:,0]) #sample standard error
    #create 99% confidence interval for the population mean
    confidenceInterval_V1 = st.norm.interval(alpha=confidenceLevel,loc=sampleMean_V1,scale=sampleStandardError_V1)

    model_results['CI_99_V1'] = [confidenceInterval_V1]
    model_results['CI_99_V1_min'] = confidenceInterval_V1[0]
    model_results['CI_99_V1_max'] = confidenceInterval_V1[1]

    sampleMean_V2 = np.mean(X_train.iloc[:,1]) #sample mean
    sampleStandardError_V2 = st.sem(X_train.iloc[:,1]) #sample standard error
    #create 99% confidence interval for the population mean
    confidenceInterval_V2 = st.norm.interval(alpha=confidenceLevel,loc=sampleMean_V2,scale=sampleStandardError_V2)

    model_results['CI_99_V2'] = [confidenceInterval_V2]
    model_results['CI_99_V2_min'] = confidenceInterval_V2[0]
    model_results['CI_99_V2_max'] = confidenceInterval_V2[1]
    
    model_results["r_sq"] = [est2.rsquared]
    model_results["adj_r_sq"] = [est2.rsquared_adj]
    y_predict_train = est2.predict(X2_train)
    y_predict_test = est2.predict(X2_test)
    mse_train = metrics.mean_squared_error(train_y, y_predict_train)
    rmse_train = np.sqrt(mse_train)
    mse_test = metrics.mean_squared_error(test_y, y_predict_test)
    rmse_test = np.sqrt(mse_test)
    mape_train = metrics.mean_absolute_percentage_error(train_y, y_predict_train)
    mape_test = metrics.mean_absolute_percentage_error(test_y, y_predict_test)
    model_results["RMSE_Dev"] = [rmse_train]
    model_results["RMSE_Val"] = [rmse_test]
    model_results["MAPE_Dev"] = [mape_train]
    model_results["MAPE_Val"] = [mape_test]

   # Shapiro Wilk Test - Normality
    res = stats.shapiro(est2.resid)
    model_results["Shapiro_Test_Statistic"] = [res.statistic]
    model_results["Shapiro_pvalue"] = [res.pvalue]

   # BP Test - Heteroskedasticity
    test = sms.het_breuschpagan(est2.resid, est2.model.exog)
    model_results["BP_Test_Statistic"] = [test[0]]
    model_results["BP_pvalue"] = [test[1]]

    # LB Test - Autocorrelation
    #perform Ljung-Box test on residuals with lag=5
    #lb_test =sm.stats.acorr_ljungbox(est2.resid, lags=[5], return_df=True)
    lb_test =sm.stats.acorr_ljungbox(est2.resid, return_df=True)
    lb_test.reset_index(drop= True)
    model_results["LB_Test_Statistic"] = lb_test.iloc[0,0]
    model_results["LB_pvalue"] = lb_test.iloc[0,1]

    return model_results
```

```python
def vif(benchmark_dataset, var_list):
    vif_output = pd.DataFrame()
    temp=var_list
    def calculate_vif(df, var_no):
        temp_vif=pd.DataFrame()
        temp_vif['variable_name']=[var_list[var_no]]
        temp_vif['VIF']=variance_inflation_factor(add_constant(df[var_list]).values, var_no+1)
        return temp_vif
    for i in range(len(var_list)):
        temp_vif = calculate_vif(benchmark_dataset[temp], i)
        vif_output=vif_output.append(temp_vif)
            
    return vif_output.sort_values('VIF').reset_index(drop=True)
```

```python
def multivariate_test_3(train_df,test_df,train_y,test_y,var1,var2,var3):
    X_train = train_df[[var1,var2,var3]]
    X_test = test_df[[var1,var2,var3]]
    X2_train = sm.add_constant(X_train)
    X2_test = sm.add_constant(X_test)
    est = sm.OLS(train_y, X2_train)
    est2 = est.fit()
    model_results = pd.DataFrame()
    model_results["Variable1"] =[var1]
    model_results["Variable2"] =[var2]
    model_results["Variable3"] =[var3]
    var_list =[var1,var2,var3]
    vif_df = vif(X_train,var_list)
    vif_df.sort_values(by='VIF',ascending= False,inplace=True)
    model_results["VIF"]=  vif_df.iloc[0,1]  # getting max of 3 VIF 
    #model_results["VIF"] = [1/(1-est2.rsquared)]
    r = X_train.corr()
#     model_results["corr_V1_V2"] = r.iloc[0,1]
#     model_results["corr_V1_V3"] = r.iloc[0,2]
#     model_results["corr_V2_V3"] = r.iloc[1,2]
    model_results["correlation"] = max(r.iloc[0,1],r.iloc[0,2],r.iloc[1,2])
    model_results["Model_Intercept"] = [est2.params[0]]
    model_results["Model_coeff_Variable1"] =[est2.params[1]]
    model_results["Model_coeff_Variable2"] = [est2.params[2]]
    model_results["Model_coeff_Variable3"] = [est2.params[3]]
    
    confidenceLevel = 0.99           #99% CI given
    degrees_freedom = len(X_train)-1 
    sampleMean_V1 = np.mean(X_train.iloc[:,0]) #sample mean
    sampleStandardError_V1 = st.sem(X_train.iloc[:,0]) #sample standard error
    #create 99% confidence interval for the population mean
    confidenceInterval_V1 = st.norm.interval(alpha=confidenceLevel,loc=sampleMean_V1,scale=sampleStandardError_V1)

    model_results['CI_99_V1'] = [confidenceInterval_V1]
    model_results['CI_99_V1_min'] = confidenceInterval_V1[0]
    model_results['CI_99_V1_max'] = confidenceInterval_V1[1]

    sampleMean_V2 = np.mean(X_train.iloc[:,1]) #sample mean
    sampleStandardError_V2 = st.sem(X_train.iloc[:,1]) #sample standard error
    #create 99% confidence interval for the population mean
    confidenceInterval_V2 = st.norm.interval(alpha=confidenceLevel,loc=sampleMean_V2,scale=sampleStandardError_V2)

    model_results['CI_99_V2'] = [confidenceInterval_V2]
    model_results['CI_99_V2_min'] = confidenceInterval_V2[0]
    model_results['CI_99_V2_max'] = confidenceInterval_V2[1]

    sampleMean_V3 = np.mean(X_train.iloc[:,2]) #sample mean
    sampleStandardError_V3 = st.sem(X_train.iloc[:,2]) #sample standard error
    #create 99% confidence interval for the population mean
    confidenceInterval_V3 = st.norm.interval(alpha=confidenceLevel,loc=sampleMean_V3,scale=sampleStandardError_V3)

    model_results['CI_99_V3'] = [confidenceInterval_V3]
    model_results['CI_99_V3_min'] = confidenceInterval_V3[0]
    model_results['CI_99_V3_max'] = confidenceInterval_V3[1]
    
    model_results["r_sq"] = [est2.rsquared]
    model_results["adj_r_sq"] = [est2.rsquared_adj]
    y_predict_train = est2.predict(X2_train)
    y_predict_test = est2.predict(X2_test)
    mse_train = metrics.mean_squared_error(train_y, y_predict_train)
    rmse_train = np.sqrt(mse_train)
    mse_test = metrics.mean_squared_error(test_y, y_predict_test)
    rmse_test = np.sqrt(mse_test)
    mape_train = metrics.mean_absolute_percentage_error(train_y, y_predict_train)
    mape_test = metrics.mean_absolute_percentage_error(test_y, y_predict_test)
    model_results["RMSE_Dev"] = [rmse_train]
    model_results["RMSE_Val"] = [rmse_test]
    model_results["MAPE_Dev"] = [mape_train]
    model_results["MAPE_Val"] = [mape_test]

   # Shapiro Wilk Test - Normality
    res = stats.shapiro(est2.resid)
    model_results["Shapiro_Test_Statistic"] = [res.statistic]
    model_results["Shapiro_pvalue"] = [res.pvalue]

   # BP Test - Heteroskedasticity
    test = sms.het_breuschpagan(est2.resid, est2.model.exog)
    model_results["BP_Test_Statistic"] = [test[0]]
    model_results["BP_pvalue"] = [test[1]]

    # LB Test - Autocorrelation
    #perform Ljung-Box test on residuals with lag=5
    #lb_test =sm.stats.acorr_ljungbox(est2.resid, lags=[5], return_df=True)
    lb_test =sm.stats.acorr_ljungbox(est2.resid, return_df=True)
    lb_test.reset_index(drop= True)
    model_results["LB_Test_Statistic"] = lb_test.iloc[0,0]
    model_results["LB_pvalue"] = lb_test.iloc[0,1]

    return model_results
```
